```python
neighborhoods=['malga','sahafah','Hittin','Yasmin','Mursalat','Aqeeq','muruj','Nargis']
data=[]
for n in neighborhoods:
    #address = '{}, Riyadh, SA'.format(n)
    geolocator = Nominatim(user_agent="ny_explorer")
    #location = geolocator.geocode(address)
    geocode = lambda query: geolocator.geocode("%s, Riyadh SA" % query)
    location=geocode(n)
    try:
        latitude = location.latitude
        longitude = location.longitude
    except:
        latitude = np.nan
        longitude = np.nan
    data.append([n,latitude,longitude])
```

#### Grouping neigborhoods

```python
clustering_df=riyadh_merged_rest_final_df[['neighborhood','verified','rating','likes_count']]
clustering_df.head()
```

```python
# one hot encoding (create dummy vars)
one_hot_df=pd.get_dummies(riyadh_merged_rest_final_df[['venue_categories']],prefix="",prefix_sep="")
one_hot_df[clustering_df.columns]=clustering_df
one_hot_df.columns
# reorder columns where neighborhood is the first one
columns=[ 'neighborhood', 'verified', 'rating','likes_count'] + [col for col in  one_hot_df.columns if col not in [ 'neighborhood', 'verified', 'rating','likes_count']] 
one_hot_df=one_hot_df.loc[:,columns]
one_hot_df.head(2)
```

```python
# group categories by neighborhoods 
one_hot_grouped=one_hot_df.groupby('neighborhood').mean().reset_index()
one_hot_grouped.head()
```

```python
# we need to transform data to be in the same scale to be ready for clustering
x=one_hot_grouped.iloc[:,1:].values
#  z-score transformation (x - mean/std) values usually between -3 and 3 may be lower or upper
x=StandardScaler().fit(x).fit_transform(x)
```

#### Choose optimal K clusters

```python
from sklearn.metrics import silhouette_score
cost =[] 
sil = []
for i in range(2,8): 
    KM = KMeans( n_clusters = i) 
    KM.fit(x) 
      
    # calculates squared error 
    # for the clustered points 
    #Sum of squared distances
    cost.append(KM.inertia_)
    labels = KM.labels_
    #sil.append(silhouette_score(x, labels, metric = 'euclidean'))
#plot the cost against K values 
plt.plot(range(2, 8), cost, color ='g', linewidth ='3') 
plt.xlabel("Value of K") 
plt.ylabel("Sum of squared distances") 
print(sil)
plt.show() # clear the plot 
```

```python
# Clustering Neighborhood with K-Means Cluster
#k-means++ selects initial cluster centers for k-mean clustering in a smart way to speed up convergence
# n_init number of run iterations with different centroid to get best output
fig=plt.figure( figsize=(11,6))
for i in np.arange(2,8):
    kmeans_model=KMeans(init = "k-means++", n_clusters = i, n_init = 12)
    kmeans_model.fit(x)
    kmeans_model.labels_
    ax = fig.add_subplot(2, 3, i-1)
    ax.scatter(x[:,0],x[:,1],c=kmeans_model.labels_)
    ax.set_title('K='+str(i))
```

It is clear that the best grouping is with K=3 or with three clusters

```python
# Apply KMEANS Algorithm with (K=3)
kmeans_model=KMeans(init = "k-means++", n_clusters = 3, n_init = 12)
kmeans_model.fit(x)
kmeans_model.labels_
```

```python
# assign labels to Neighborhoods
one_hot_grouped['label']=kmeans_model.labels_
one_hot_grouped_lebeled=one_hot_grouped[['neighborhood','label']]
one_hot_grouped_lebeled.head()
```

```python
from folium import plugins
riyadh_map = folium.Map(location=['24.7136', '46.6753'], zoom_start=10.5)#Riyadh Coordinates
# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(riyadh_map)

# group venues by neighborhoods in the map
for lat, lng, neigh,name,cat in zip(gr0.venue_lat, gr0.venue_lng, gr0.neighborhood, gr0.venue_name, gr0.venue_categories):
    label='<b>venue</b>:{}<br/><b>category</b>:{}<br/><b>neigborhood</b>:{}'.format(name,cat,neigh)
    iframe = folium.IFrame(html=label, width=220, height=100)
    label_popup = folium.Popup(iframe, max_width=2650)
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label_popup,
    ).add_to(incidents)

# display map
riyadh_map
```

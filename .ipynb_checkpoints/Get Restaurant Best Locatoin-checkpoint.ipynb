#### Get coordinates for each Neigborhood

```python
neighborhoods=['malga','sahafah','Hittin','Yasmin','Mursalat','Aqeeq','muruj','Nargis']
data=[]
for n in neighborhoods:
    #address = '{}, Riyadh, SA'.format(n)
    geolocator = Nominatim(user_agent="ny_explorer")
    #location = geolocator.geocode(address)
    geocode = lambda query: geolocator.geocode("%s, Riyadh SA" % query)
    location=geocode(n)
    try:
        latitude = location.latitude
        longitude = location.longitude
    except:
        latitude = np.nan
        longitude = np.nan
    data.append([n,latitude,longitude])
```

```python
# extract categroy from json
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
venue_df['categories']=venue_df.apply(get_category_type,axis=1)
venue_df.head(3)
```

```python
# remove duplicates
venue_df=venue_df[venue_df.id.duplicated()==False]
#check
venue_df.id.duplicated().sum(),venue_df.shape
```

```python
#reorder columns
filtered_columns = ['Neighborhood','id','name', 'categories','hasPerk'] + [col for col in venue_df.columns if col.startswith('location.')]
venue_df=venue_df.loc[:,filtered_columns]
```

```python
# rename columns
dic={}
for col in venue_df:
    if col.startswith('location.'):
        dic[col]=col.replace('location.','venue_')
        #dic= str(dic) + "'{}':'{}',".format(col,col.replace('location.','venue_'))
    else:
         dic[col]='venue_'+str(col)
        #dic=str(dic) +"'{}':'{}',".format(col,'venue_'+str(col))
venue_df=venue_df.rename(columns=dic)
```

#### Grouping neigborhoods

```python
clustering_df=riyadh_merged_rest_final_df[['neighborhood','verified','rating','likes_count']]
clustering_df.head()
```

```python
# one hot encoding
one_hot_df=pd.get_dummies(riyadh_merged_rest_final_df[['venue_categories']],prefix="",prefix_sep="")
one_hot_df[clustering_df.columns]=clustering_df
one_hot_df.columns
#one_hot_df['neighborhood']=riyadh_merged_rest_final_df['neighborhood']
# reorder columns where neighborhood is the first one
columns=[ 'neighborhood', 'verified', 'rating','likes_count'] + [col for col in  one_hot_df.columns if col not in [ 'neighborhood', 'verified', 'rating','likes_count']] 
one_hot_df=one_hot_df.loc[:,columns]
one_hot_df.head(2)
```

```python
# group categories by neighborhoods 
one_hot_grouped=one_hot_df.groupby('neighborhood').mean().reset_index()
one_hot_grouped.head()
```

```python
# we need to transform data to be in the same scale to be ready for clustering
x=one_hot_grouped.iloc[:,1:].values
#  z-score transformation (x - mean/std) values usually between -3 and 3 may be little lower or upper
x=StandardScaler().fit(x).fit_transform(x)
```

```python
# Clustering Neighborhood with K-Means Cluster
#k-means++ selects initial cluster centers for k-mean clustering in a smart way to speed up convergence
# n_init number of run iterations with different centroid to get best output
kmeans_model=KMeans(init = "k-means++", n_clusters = 3, n_init = 12)
kmeans_model.fit(x)
kmeans_model.labels_
```

```python
# assign labels to Neighborhoods
one_hot_grouped['label']=kmeans_model.labels_
one_hot_grouped_lebeled=one_hot_grouped[['neighborhood','label']]
one_hot_grouped_lebeled.head()
```

```python
# lets add the labels to or main data frame
riyadh_merged_rest_final_df_labled=pd.merge(riyadh_merged_rest_final_df,one_hot_grouped_lebeled,left_on='neighborhood',right_on='neighborhood')
riyadh_merged_rest_final_df_labled.head(2)
```
